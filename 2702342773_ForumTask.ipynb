```python
# Forum Task 1

lloyd = {
    "name": "Lloyd",
    "homework": [90.0, 97.0, 75.0, 92.0],
    "quizzes":[88.0, 40.0, 94.0],
    "tests": [75.0, 90.0]
}
alice = {
    "name": "Alice",
    "homework": [100.0, 92.0, 98.0, 100.0],
    "quizzes": [82.0, 83.0, 91.0],
    "tests": [89.0, 97.0]
}
tyler = {
    "name":"Tyler",
    "homework": [0.0, 87.0, 75.0, 22.0],
    "quizzes": [0.0, 75.0, 78.0],
    "tests": [100.0, 100.0]
}

students = [lloyd, alice, tyler]  

for l in lloyd:
    lloyds = lloyd.get(l)
    print(lloyds)
for a in alice:
    alices = alice.get(a)
    print(alices)
for t in tyler:
    tylers = tyler.get(t)
    print(tylers)

def average(numbers):
    total = sum(numbers)
    total = float(total)
    return total/len(numbers)
for s in students:
    print(average(s["homework"]))
    
    
def get_average(student):
    homework = average(student["homework"])
    quizzes = average(student["quizzes"])
    tests = average(student["tests"])
    homeworkWeight = homework * 0.1
    quizzesWeight = quizzes * 0.3
    testsWeight = tests * 0.6
    totalWeight = homeworkWeight + quizzesWeight + testsWeight
    return totalWeight


def get_letter_grade(score):
    if score >= 90.0:
        return "A"
    elif score >= 80.0:
        return "B"
    elif score >= 70.0:
        return "C"
    elif score >= 60.0:
        return "D"
    else:
        return "F"
print(get_letter_grade(get_average(lloyd)))

def get_class_average(students):
    results = []
    for s in students:
        getAverage=get_average(s)
        results.append(getAverage)
        return average(results)
print(get_class_average(students))

print(get_letter_grade(get_class_average(students)))
```

```python
# Forum Task 3
daysOfTheWeek = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]

sales=[]

for days in daysOfTheWeek:
    
    quest = eval(input(f"Enter the sales for {days}: "))
    
    if quest < 0:
        input(f"Input was invalid. Re-enter the sales for {days}: ")
    
        
    sales.append(quest)
    print(sales)


totalSales=print(f"The total sales is {sum(sales)}")
minAndMax = print(f"The minimum sales amount was: {min(sales):.2f} and the maximum amount was: {max(sales):.2f}")
```
